```python
import pymp

first_page = 1
last_page = 167
with pymp.Parallel(4) as p:
    for index in p.range(first_page, last_page):
        meigen = requests.get("http://number.bunshun.jp/category/meigen?page={}".format(index))
        soup = BeautifulSoup(meigen.content, "html.parser")
        section = soup.find("div", {"class": "saying-list"})
        urls = [li.find("a")['href'] for li in section.find_all("li")]
        with open("urllist.txt", "a") as f:
            f.write('\n'.join(urls) + "\n")
        p.print('{} done!'.format(index))
```

```python
url = "https://ja.wikiquote.org/wiki/%E9%BA%BB%E5%8E%9F%E5%BD%B0%E6%99%83"
r = requests.get(url)
soup = BeautifulSoup(r.content, "html.parser")
section = soup.find("div",{"class": "mw-parser-output"})
[d.extract() for d in section.find_all("ol")]
[d.extract() for d in section.find_all("dl")]
section.find("div", {"id": "toc"}).extract()
person, quotes = soup.find("h1",{"id": "firstHeading"}).get_text(), [re.sub(r'\[[0-9]+\]',"", d.get_text().replace("「","").replace("」","").replace("\u3000","").replace("\n", "").strip()) for d in section.find_all("li")]
```

```python
url = "http://www.meigensyu.com/quotations/index/page{}.html"

with open("quote_list_large.tsv", "w") as f:
    f.write("quote\tperson\n")

with pymp.Parallel(100) as p:
    for index in p.range(0, 98):
        try:
            r = requests.get(url.format(index+1))
            soup = BeautifulSoup(r.content, "html.parser")
            contents = soup.find_all("div", {"class": "meigenbox"})
            for content in contents:
                link = content.find("div", {"class": "link"})
                lis = link.find_all("li")
                person = lis[0].get_text()
                if "出典" in person:
                    person = re.sub(r'.+?:(.*)', r'\1', lis[1].get_text()).strip()

                if not person:
                    continue
                quote = content.find("div", {"class": "text"})
                [d.extract() for d in quote.find_all("div",{"class": "en_text"})]
                quote = quote.get_text().strip()
                with open("quote_list_large.tsv", "a") as f:
                    f.write("{}\t{}\n".format(quote, person))
            p.print('{} done!'.format(index))
        except:
            print("ERROR: {}".format(index))
```

```python
with open("urllist_mkeizi.txt", "r") as f:
    urllist = [d.strip() for d in f.readlines()]
```

```python
url = urllist[0]
r = requests.get(url)
soup = BeautifulSoup(r.content, "html.parser")
persons = soup.find_all("div", {"class":"wpcr_fl wpcr_rname"})
quotes = soup.find_all("div", {"class": "description"})

persons = [person.get_text() for person in persons]
quotes = [quote.get_text() for quote in quotes]

assert len(persons) == len(quotes)
```

```python
with open("quote_list_mkeizi.tsv", "w") as f:
    f.write("quote\tperson\n")

with pymp.Parallel(1000) as p:
    for index in p.range(0, len(urllist)):
        try:
            url = urllist[index]
            r = requests.get(url)
            soup = BeautifulSoup(r.content, "html.parser")
            persons = soup.find_all("div", {"class":"wpcr_fl wpcr_rname"})
            quotes = soup.find_all("div", {"class": "description"})

            persons = [person.get_text() for person in persons]
            quotes = [quote.get_text() for quote in quotes]

            assert len(persons) == len(quotes)
            
            for quote, person in zip(quotes, persons):
                with open("quote_list_mkeizi.tsv", "a") as f:
                    f.write("{}\t{}\n".format(quote, person))
            
            p.print('{} done!'.format(index))
        except:
            print("ERROR: {}".format(index))
```

```python
with open("quote_list_kotodama.tsv", "w") as f:
    f.write("quote\tperson\n")

with pymp.Parallel(100) as p:
    for index in p.range(0, 769):
        try:
            url = "http://www.kotodama.in/bbs/board/remark/{}".format(index+1)
            r = requests.get(url)
            soup = BeautifulSoup(r.content, "html.parser")
            quotes = [d.get_text().replace('\n', '').replace("\xa0","") for d in soup.find_all("td")]
            quotes = [re.sub(r'^.+?：','', d) for d in quotes if len(list(d)) > 0 and list(d)[0].isnumeric()]
            persons = [d.get_text() for d in soup.find_all("td", {"width": "120"})][1:]

            assert len(quotes) == len(persons)
            
            for quote, person in zip(quotes, persons):
                with open("quote_list_kotodama.tsv", "a") as f:
                    f.write("{}\t{}\n".format(quote, person))
            
            p.print('{} done!'.format(index))
        except Exception as e:
            print("ERROR: {}, {}".format(index, e))
```
